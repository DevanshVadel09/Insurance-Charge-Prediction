## Missing Data Handling 

```python
# Drpo the charges 
df = df.dropna(subset=['charges'])
# fill missing value
df['bmi'] = df['bmi'].fillna(df['bmi'].mean())
df['smoker'] = df['smoker'].fillna('Unknown') # it was not in numbers
```

## Visualize Relationships

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='smoker', y='charges', data=df)     # Smokers vs charges
sns.scatterplot(x='bmi', y='charges', data=df)    # BMI vs charges
sns.lmplot(x='age', y='charges', hue='smoker', data=df)  # Age vs charges by smoker
```
